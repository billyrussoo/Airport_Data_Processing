```python
import pandas as pd
from io import StringIO

with open('Airport_Locations.txt', 'r', encoding='utf-8') as file:
    data = file.read().replace(r'\N', '')
df = pd.read_csv(StringIO(data), header=None, names=[
    'AirportID', 'Name', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 
    'Altitude', 'Timezone', 'DST', 'TzDatabaseTimeZone', 'Type', 'Source'
])

print(df[['IATA', 'Latitude', 'Longitude']])
df.to_csv('output.csv', index=False)

```

```python
import pandas as pd

df = pd.read_csv('output.csv')
df_us = df[df['Country'] == 'United States']
print(df_us[['IATA', 'Latitude', 'Longitude']])
df_us.to_csv('Us_Airports.csv', index=False)
```

```python
import pandas as pd

df_iata = pd.read_csv('flights_sample_3m.csv')
df_longitude_latitude = pd.read_csv('Us_Airports.csv')
df_merged = pd.merge(df_iata, df_longitude_latitude[['IATA', 'Latitude', 'Longitude']], left_on='ORIGIN', right_on='IATA', how='left')
print(df_merged[['ORIGIN', 'Latitude', 'Longitude']])
df_merged.to_csv('merged_output.csv', index=False)

```
