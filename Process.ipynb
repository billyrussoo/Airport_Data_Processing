```python
import pandas as pd
from io import StringIO

with open('Airport_Locations.txt', 'r', encoding='utf-8') as file:
    data = file.read().replace(r'\N', '')
df = pd.read_csv(StringIO(data), header=None, names=[
    'AirportID', 'Name', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 
    'Altitude', 'Timezone', 'DST', 'TzDatabaseTimeZone', 'Type', 'Source'
])

print(df[['IATA', 'Latitude', 'Longitude']])
df.to_csv('output.csv', index=False)

```

```python
import pandas as pd

df_iata = pd.read_csv('flights_sample_3m.csv')
df_longitude_latitude = pd.read_csv('Us_Airports.csv')

df_merged_origin = pd.merge(df_iata, df_longitude_latitude[['IATA', 'Latitude', 'Longitude']], left_on='ORIGIN', right_on='IATA', how='left')
df_merged_origin.rename(columns={'Latitude': 'o_latitude', 'Longitude': 'o_longitude'}, inplace=True)

df_merged_dest = pd.merge(df_merged_origin, df_longitude_latitude[['IATA', 'Latitude', 'Longitude']], left_on='DEST', right_on='IATA', how='left')
df_merged_dest.rename(columns={'Latitude': 'd_latitude', 'Longitude': 'd_longitude'}, inplace=True)

df_merged_dest.drop(columns=['IATA_x', 'IATA_y'], inplace=True)


print(df_merged_dest[['ORIGIN', 'o_latitude', 'o_longitude', 'DEST', 'd_latitude', 'd_longitude']])


df_merged_dest.to_csv('Longitues_flights_merged.csv', index=False)
```

```python
import pandas as pd
df = pd.read_csv('Longitues_flights_merged.csv')
iata_coords = {
    'STX': (17.698420, -64.798600),
    'STT': (18.334470, -64.971260),
    'BQN': (18.488775, -67.130653),
    'SJU': (18.425578, -66.104856),
    'PSE': (18.014440, -66.564020),
    'SPN': (15.120491, 145.729625),
    'GUM': (13.385870, 144.670430),
    'EAR': (40.725630, -99.013060),
    'PPG': (-14.215810, -169.425560),
    'XWA': (48.259722222, -103.7505555556)
}

def update_lat_lon(df, iata_coords, col_prefix, lat_col, lon_col):
    for iata, coords in iata_coords.items():
        mask = (df[col_prefix] == iata) & (df[lat_col].isnull())
        df.loc[mask, [lat_col, lon_col]] = coords
update_lat_lon(df, iata_coords, 'ORIGIN', 'o_latitude', 'o_longitude')
update_lat_lon(df, iata_coords, 'DEST', 'd_latitude', 'd_longitude')

df.to_csv('Final_FLights_Cords.csv', index=False)

print(df[['ORIGIN', 'o_latitude', 'o_longitude', 'DEST', 'd_latitude', 'd_longitude']])

```
